```python
df.drop("education",axis=1,inplace=True)
```

```python
df.columns
```

```python
skewed = ['capital-gain', 'capital-loss']
df[skewed] = df[skewed].apply(lambda x: np.log(x + 1))
df.head()
```

```python
df.hist(column="capital-loss",bins=20)

df.hist(column="capital-gain",bins=20)
```

```python
df = df[df["workclass"]!="?"]
df = df[df["occupation"] != "?"]
df = df[df["native-country"]!="?"]
```

```python
df.shape
```

```python
df.drop("fnlwgt",axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
num_col = ["age","education-num","capital-gain","capital-loss","hours-per-week"]
scaler = MinMaxScaler()
# make a copy of the original data frame
min_max_df = pd.DataFrame(data=df)
min_max_df[num_col] = scaler.fit_transform(min_max_df[num_col])
min_max_df.head()
```

```python
# Encode the 'income_raw' data to numerical values
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(min_max_df["Income"])
```

```python
le.classes_
```

```python
min_max_df["Income"] = le.transform(min_max_df["Income"])
```

```python
min_max_df.head()
```

```python
y = min_max_df["Income"]
X = min_max_df.drop("Income",axis=1)
```

```python
X = pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.shape , y_train.shape
```

```python
X_test.shape , y_test.shape
```

```python
print((len(y_train[y_train == 0])*100)/22792,(len(y_test[y_test == 0])*100)/9769)
```

```python
print((len(y_train[y_train == 1])*100)/22792,(len(y_test[y_test == 1])*100)/9769)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
print(clf.score(X_train,y_train))
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_clf=accuracy_score(y_test,y_pred)*100
```

```python
accuracy_clf
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'min_samples_split':range(10,500,20),'max_depth':range(1,20,2),'criterion':['gini','entropy']}
clf_tree=DecisionTreeClassifier()
clf=GridSearchCV(clf_tree,parameters,cv=5)
clf.fit(X_train,y_train)
```

```python
clf.best_estimator_
```

```python
clf_predict=clf.predict(X_test)
```

```python
accuracy_score(y_test,clf_predict)
```
